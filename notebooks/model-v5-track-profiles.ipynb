```python
import fastf1
fastf1.Cache.enable_cache("../data/cache")

schedule_2025 = fastf1.get_event_schedule(2025)

all_races_2025 = []

for _, event in schedule_2025.iterrows():
    try:
        race_name = event["EventName"]
        session = fastf1.get_session(2025, race_name, "R")
        session.load()

        race_results = session.results[[
            "Abbreviation",
            "TeamName",
            "GridPosition",
            "Position",
            "Status"
        ]].copy()

        race_results["RaceName"] = race_name
        race_results["RoundNumber"] = event["RoundNumber"]
        race_results["Year"] = 2025

        all_races_2025.append(race_results)

        print(f"Successfully loaded {race_name}")
    except Exception as e:
        print(f"Error loading {race_name}: {e}")

df_2025 = pd.concat(all_races_2025, ignore_index=True)

save_dataset(df_2025, "../data/processed/race_results_2025.csv")
```
